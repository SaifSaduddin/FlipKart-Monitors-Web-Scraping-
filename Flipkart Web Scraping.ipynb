```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests

load_webpage = requests.get('https://www.flipkart.com/desktop-pc-store?otracker=nmenu_sub_Electronics_0_Desktop%20PCs&otracker=nmenu_sub_Electronics_0_Desktop%20PCs&otracker=nmenu_sub_Electronics_0_Desktop%20PCs&otracker=nmenu_sub_Electronics_0_Desktop%20PCs')

Soup_object = bs(load_webpage.content)
HTML = Soup_object.prettify()
info_box = Soup_object.find_all(class_='_4ddWXP')

box = info_box[2]

TITLE = box.find(class_ = 's1Q9rs').get_text()
PRICE = box.find(class_ = '_30jeq3').get_text()
RATING = box.find(class_ = '_3LWZlK').get_text()

# print(Price)
# print(Ratings)

data_list = []
for box in info_box:
    try:
        initial_dictionary = {}
        initial_dictionary['TITLE'] = box.find(class_ = 's1Q9rs').get_text()
        initial_dictionary['PRICE'] = box.find(class_ = '_30jeq3').get_text().replace('â‚¹', '')
        initial_dictionary['RATING'] = box.find(class_ = '_3LWZlK').get_text()
        data_list.append(initial_dictionary)
    
    except:
        continue
data_list
```

```python
df = pd.DataFrame(data_list)
df
```
